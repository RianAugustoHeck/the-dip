```python
import skimage as ski
from matplotlib import pyplot as pl
import numpy as np

imgs = []
items = []
for folder in path_imgs.iterdir():
  for item in folder.iterdir():
    items.append(item)

items = sorted(items)

for item in items:
    img = ski.io.imread(item)
    imgs.append(img)
```

```python
mont_capacitores = ski.util.montage(imgs[0:4], channel_axis=-1, grid_shape=(1, 4))
mont_leds = ski.util.montage(imgs[4:8], channel_axis=-1, grid_shape=(1, 4))
mont_resistores = ski.util.montage(imgs[8:12], channel_axis=-1, grid_shape=(1, 4))
mont_seven_seg = ski.util.montage(imgs[12:16], channel_axis=-1, grid_shape=(1, 4))
mont_transist = ski.util.montage(imgs[16:20], channel_axis=-1, grid_shape=(1, 4))
```

```python
fig, axes = pl.subplots(5, 1, figsize=(40.96,20.48))

ordem = [mont_capacitores, mont_leds, mont_resistores, mont_seven_seg, mont_transist]
nomes = ['Capacitor', 'Led', 'Resistor', 'Display de sete segmentos', 'Transistor']
for i, mont in zip(range(0,5), ordem):
  axes[i].imshow(mont)
  axes[i].set_axis_off()
  axes[i].set_title(nomes[i])
```

# Exibir metadados

```python
from ast import Pass
import json


def imprimir(item):
  if isinstance(item, list):
    while len(item) != 0:
      imprimir(item[0])
      item.pop(0)
  elif isinstance(item, dict):
    chaves = item.keys()
    for chave in chaves:
      print(chave + ' : ', end='')
      imprimir(item[chave])
  elif isinstance(item, str):
    print(item)

def exibir_metadados():
  with open (path_root.joinpath(pathlib.Path('metadata.json'))) as arquivo:
    md = json.load(arquivo)
    for item in md:
      print(item.upper() + ':')
      imprimir(md[item])
      print('\n' + '--'*10 + '\n')

exibir_metadados()
```
